```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("/home/hammadali08/Vidoes/CSV file/Salary Data.csv")
data.drop(columns=['Gender','Education Level','Job Title'],inplace=True)
data
```

```python
data.fillna(data.mean(),inplace=True)
data['Salary'].astype("int64")
```

```python
sns.scatterplot(x='Age',y='Years of Experience',data=data,hue='Salary')
plt.show()
```

```python
x=data.iloc[:,:-1]
y=data['Salary']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knr=KNeighborsRegressor(n_neighbors=4)
```

```python
knr.fit(x_train,y_train)
```

```python
knr.score(x_test,y_test),knr.score(x_train,y_train)
```

```python
for i in range(1,20):
    knr2=KNeighborsRegressor(n_neighbors=i)
    knr2.fit(x_train,y_train)
    print(i,knr2.score(x_test,y_test),knr2.score(x_train,y_train))
```
