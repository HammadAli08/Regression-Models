```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data=pd.read_csv("/home/hammadali08/Videos/CSV file/cardata.csv")
data.head()
```

```python
x=data[['Present_Price']]
y=data['Selling_Price']
```

```python
plt.scatter(x,y)
plt.xlabel("Present Price")
plt.ylabel("Selling Price")
plt.show()
```

Let's do train test split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=70)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lg=LinearRegression()
lg.fit(x_train,y_train)
```

```python
lg.predict([[5.59]])
```

```python
lg.score(x_test,y_test)
```

```python
plt.scatter(x,y,label="Original Data")
plt.xlabel("Present Price")
plt.ylabel("Selling Price")
plt.plot(x,lg.predict(x),color="yellow",label="Predicted Data")
plt.legend()
plt.show()
```
