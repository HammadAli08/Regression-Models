```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("/home/hammadali08/Vidoes/CSV file/Salary Data.csv")
data.drop(columns=['Gender','Education Level','Job Title'],inplace=True)
data
```

```python
data.fillna(data.mean(),inplace=True)
data['Salary']=pd.DataFrame(data['Salary'].astype('int64'))
```

```python
sns.scatterplot(x="Years of Experience",y="Age",hue="Salary",data=data)
plt.show()
```

```python
x=data.iloc[:,:-1]
y=data['Salary']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=50)
```

```python
from sklearn.svm import SVR
```

```python
sv=SVR(kernel='linear')
```

```python
sv.fit(x_train,y_train)
```

```python
sv.score(x_test,y_test)
```
